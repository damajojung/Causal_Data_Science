# D-separation game

## D-separation definition:
A path between i and j is **blocked by Z** (a subset of all other variabels) if **at least one** condition holds:
- There is a non-collider l on the path that is in **Z**, i.e. $l \in \mathbf{Z}$, **OR**
- There is a collider k on the path, i.e. $\to k \leftarrow$, but none of its descendants (including itself) are in **Z**, i.e. $Desc(k) \cap \mathbf{Z} = \emptyset$

We say i and j are **d-separated by Z** if all paths between them are blocked by Z.

## Game introduction:
This game has two players: you and an automatic adversary.
- You are trying to remove nodes to create d-separation between two given nodes.
- The adversary is trying to add nodes to create d-connections between two given nodes.

The game starts with a random graph of size num_nodes=5 with some red nodes, x and y, that you are trying to d-separate by removing k=1 nodes from a conditioning set, z, which is denoted with green nodes.
- You have the first turn and you need to remove k nodes in order to create a d-separation (this is always possible). 
-- If you don't remove the correct nodes and the red nodes are still d-connected given the new green nodes, you **lose** one of your three lives.
-- If you lose all the lives, you have to start again the game from zero.
-- If you remove the correct nodes for three different graphs, you will **level up**: the number of nodes in the new graph will increase by one.
- The adversary then randomly selects a set of k nodes (none of them can be the same as the ones you removed) to add to the conditioning set. 
-- If the new green nodes added by the adversary create a d-connection, then you **draw** and add 1 to your score.
-- If the adversary chooses nodes that do not create a d-connection, you **win** and add num_nodes * k to your score.

***Attention***
Both num_nodes and k are configurable, but the graph generation gets very slow above num_nodes = 8, so I put that as a limit.
Similarly the generation of an appropriate graph gets very slow for k>2.

You can start the game by calling:
`g = DsepGame()` and you can play your turn by choosing which nodes to remove from the conditioning set by calling `g.play([3,4])`. If you want to play a new game but keep the score and the levels you can call `g.newGame()`
