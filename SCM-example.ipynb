```python
import numpy as np
from numpy.random import randn
import pandas as pd
import seaborn as sns

# We need a lot of samples to plot the conditional distribution:
n_samples=100000
```

We simulate samples from the SCM from the example X = \epsilon_X ; Y = 4 * X + \epsilon_Y:

```python
x = randn(n_samples)
y = 4 * x + randn(n_samples)
# plot P(X,Y)
sns.scatterplot(x=x,y=y)
```

```python
# plot P(X)
# If seaborn doesn't work use histograms: pyplot.hist(x) from matplotlib
## Estimated density plot: sns.kdeplot(x)
sns.displot(x)
```

```python
# plot P(Y)
sns.displot(y)
```

We simulate data from an intervention setting X to 2:

```python
x_do_x = np.array([2] * n_samples, dtype="int32")
y_do_x = 4 * x_do_x  + randn(n_samples)

# plot P(X,Y |do(X=2))
sns.scatterplot(x=x_do_x,y=y_do_x)
```

```python
# plot P(Y|do(X=2))
#sns.kdeplot(y_do_x)
sns.displot(y_do_x)
```

```python
# plot P(Y|X=2) (we consider +/-0.1 to have enough data:
df = pd.DataFrame({"x": x, "y": y })
df_cond_x = df[(df['x'] <= 2+0.1) & (df['x'] >= 2-0.1)]
sns.displot(df_cond_x['y'])
```

We now show what happens when you intervene on Y setting it to 4:

```python
x_do_y = randn(n_samples)
y_do_y = 4

# plot P(X,Y |do(Y=4))
sns.scatterplot(x=x_do_y,y=y_do_y)
```

```python
# plot P(X|do(Y=4))
sns.displot(x_do_y)
```

```python
# plot P(X)
sns.displot(x)
```

```python
df_cond_y = df[(df['y'] <= 4+0.1) & (df['y'] >=4-0.1)]
sns.displot(df_cond_y['x'])
```
